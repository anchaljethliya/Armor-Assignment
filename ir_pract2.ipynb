```python
# 2a. Inverted index with linguistic preprocessing

import re
from collections import defaultdict
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# Initialize stopwords and stemmer
import nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
```

```python
# Documents collection
documents = {
    "D1": "Information Retrieval system",
    "D2": "neural networks to AI-powered conversations",
    "D3": "dives deep into the ethical considerations that are shaping the future of IR",
    "D4": "Information retrieval system design",
    "D5": "hello",
    "D6": "my name is Anchal",
    "D7": "information retrieval and search engines",
    "D8": "retrieval models in information retrieval",
    "D9": "indexing and searching techniques",
    "D10": "introduction to information systems"
}
```

```python
def normalize(text):
    # Tokenize and lowercase
    tokens = re.findall(r'\w+', text.lower())  # matches [a-zA-Z0-9_]

    # Stopword removal and stemming
    normalized_terms = [
        stemmer.stem(token)
        for token in tokens
        if token not in stop_words
    ]

    return normalized_terms

def inverted_index(docs):
    index = defaultdict(set)

    # Process documents in sorted order
    for doc_id in sorted(docs.keys()):
        terms = normalize(docs[doc_id])
        for term in terms:
            index[term].add(doc_id)

    # Sort terms and posting lists
    sorted_index = {
        term: sorted(index[term])
        for term in sorted(index.keys())
    }

    return sorted_index

# Build inverted index
sort_inverted_index = inverted_index(documents)

# Display results
print("Sorted Inverted Index (with Linguistic preprocessing):")
for term, postings in sort_inverted_index.items():
    print(f"{term} → {postings}")

print("---------- Dictionary size --------")
print(f"Vocabulary Size: {len(sort_inverted_index)}")

print("---------- Total Document Indexed --------")
print(f"Total document: {len(documents.keys())}")
```

```python
# 2b. Inverted Index on Folder of Text Files

import os
import re
from collections import defaultdict
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# Create folder to store documents
os.makedirs("docs", exist_ok=True)

# Upload files (Google Colab only)
from google.colab import files
uploaded = files.upload()

# Move uploaded files into docs folder
for filename in uploaded.keys():
    os.rename(filename, os.path.join("docs", filename))
```

```python
# Initialize NLP tools
import nltk
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()

def preprocess(text):
    # Extract only alphabetic tokens and lowercase
    tokens = re.findall(r'\b[a-zA-Z]+\b', text.lower())

    # Stopword removal and lemmatization
    processed = [
        wordnet_lemmatizer.lemmatize(token)
        for token in tokens
        if token not in stop_words
    ]

    return processed
```

```python
def folder_inverted_index(folder_path):
    inverted_index = defaultdict(set)
    all_docs = set()
    count_document = 0

    for filename in sorted(os.listdir(folder_path)):
        if filename.endswith(".txt"):
            count_document += 1
            doc_id = filename
            all_docs.add(doc_id)

            file_path = os.path.join(folder_path, filename)
            with open(file_path, "r", encoding="latin-1") as file:
                text = file.read()

            terms = preprocess(text)
            for term in terms:
                inverted_index[term].add(doc_id)

    # Sort vocabulary and posting lists
    sorted_index = {
        term: sorted(inverted_index[term])
        for term in sorted(inverted_index.keys())
    }

    return sorted_index, all_docs, count_document
```

```python
# -------- Execute Program --------
folder_path = "docs"
sorted_index, all_docs, total_docs = folder_inverted_index(folder_path)

print("Sorted Inverted Index (with linguistic preprocessing):\n")
for term, postings in sorted_index.items():
    print(f"{term} → {postings}")

print("\n---------- Dictionary Size --------")
print(f"Vocabulary Size: {len(sorted_index)}")

print("\n---------- Total Documents Indexed --------")
print(f"Total documents indexed: {total_docs}")
```
